```python
with open("tarot-json/tarot-images.json", "r") as j:
    tarot_json = json.load(j)
```

```python
df = pd.DataFrame.from_dict(card_dict)
```

```python
spread = []
```

```python
while len(spread)<=10:
    drawn = int(quantumrandom.randint(0,77))
    if drawn not in spread:
        spread.append(drawn)
```

```python
spread
```

```python
spread_impaths = [str("tarot-json/cards/" + df.iloc[spread_ind]["img"]) for spread_ind in spread]
```

```python
spread_impaths
```

```python
from IPython.display import Image, display

for impath in spread_impaths:
    display(Image(filename=impath))
```
